```python
import pandas as pd
```

```python
df = pd.read_csv("base_rh.csv", sep=';', encoding='cp1252', decimal=',')
df.head(5)
df.tail()
df.info()
df.describe()
df.nunique()

#df.['Col'].value_counts()
df.isnull().sum()
df.dropna(inplace=True) # remove linhas com valores nulos
#df.fillna(df.mean(), inplace=True) # preencher valores nulos com a média da coluna

df['Nome']
df[['Nome', 'Idade']]
df[df['Status'] == 'Ativo']
df[(df['Status'] == 'Ativo') & (df['Idade'] > 30)]

df['Data de Admissão'] = pd.to_datetime(df['Data de Admissão'], format='%d/%m/%Y', errors='coerce')
df['Ano'] = df['Data de Admissão'].dt.year
df['Mês'] = df['Data de Admissão'].dt.month
df['Trim'] = df['Data de Admissão'].dt.quarter
df['Tempo'] = (pd.to_datetime('today') - df['Data de Admissão']).dt.days / 365.25.__round__(2)

novos = df[df['Data de Admissão'] >= pd.Timestamp.today() - pd.DateOffset(years=1)]
antigos = df[df['Data de Admissão'] < pd.Timestamp.today() - pd.DateOffset(years=1)]

df.groupby('Departamento')['Salário'].mean().__round__(2)
df.groupby('Departamento')['Salário'].mean().sort_values(ascending=False)

# groupby para mostrar maior e menor tempo de empresa por departamento
df.groupby('Departamento')['Tempo'].agg(['min', 'max'])
```
