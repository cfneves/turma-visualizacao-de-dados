Aula_01 - Instalar as bibliotecas e carregar o csv.

```python
import pandas as pd
```

```python
# Montar o Dataframe (df) carregando o arquivo de csv diretamente pela URL.

URL = (
    "https://raw.githubusercontent.com/"
    "cfneves/turma-visualizacao-de-dados/"
    "master/aulas/semana_04/base_rh.csv"
)

# Carregar o csv com as configurações corretas para exibir os dados de forma integra.

# sep      → ponto-e-vírgula porque o TOTVS usa padrão europeu
# encoding → cp1252 é o encoding do Windows (suporta ã é ç ú ô)
# decimal  → vírgula como separador decimal

df = pd.read_csv(URL, sep=",", encoding="cp1252", decimal=",")
```

```python
# Conferência imediata — sempre faça isso após qualquer leitura

print(f"Shape: {df.shape}")
print(f"\nTipos:\n{df.dtypes}")
print(f"\nPrimeiras 3 linhas:\n{df.head(3).to_string()}")
```
