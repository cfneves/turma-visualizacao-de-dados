**Aula_01 - Instalar as bibliotecas e carregar o csv.**

```python
import pandas as pd
print(pd.__version__)
```

```python
# Montar o Dataframe (df) carregando o arquivo de csv diretamente pela URL.

URL = (
    "https://raw.githubusercontent.com/"
    "cfneves/turma-visualizacao-de-dados/"
    "master/datasets/base_rh.csv"
)

# Carregar o csv com as configurações corretas para exibir os dados de forma integra.

# sep      → ponto-e-vírgula porque o TOTVS usa padrão europeu
# encoding → cp1252 é o encoding do Windows (suporta ã é ç ú ô)
# decimal  → vírgula como separador decimal

df = pd.read_csv(URL, 
    sep=";", 
    encoding="cp1252", 
    decimal=","
    )
```

```python
# Conferência imediata — sempre faça isso após qualquer leitura

print(f"Shape: {df.shape}") #Mostra o número de linhas e colunas do DataFrame
print(f"\nTipos:\n{df.dtypes}") #Mostra os tipos de dados de cada coluna do DataFrame
print(f"\nPrimeiras 3 linhas:\n{df.head(3).to_string(index=False)}") #Mostra as primeiras 3 linhas do DataFrame sem os índices
print(f"\nÚltimas 5 linhas:\n{df.tail(5).to_string(index=False)}") #Mostra as últimas 5 linhas do DataFrame sem os índices
```

**Por que isso importa?**

Note que Salario já é float64 — isso significa que df["Salario"].mean() vai funcionar. 
Se esquecermos decimal=",", Salario vira object e a média retorna TypeError. 
Esse erro só aparece quando você tenta calcular — e já passou horas analisando o dado "errado" sem saber.

**Exploração básica do DataFrame — Antes de qualquer transformação**

Após qualquer leitura, o primeiro passo é sempre explorar o dado. Esses comandos são o "scanner" do analista — você nunca mexe num dado sem antes conhecê-lo. Execute cada um em uma célula separada no notebook para ver o resultado com clareza.

```python
# Célula 2 — shape e colunas
# Dimensão do DataFrame: (Linhas e colunas)
print("Dimensão do DataFrame (linhas, colunas):\n")

print(f"Shape: {df.shape}")      #Mostra o número de linhas e colunas do DataFrame
print(f"Linhas: {df.shape[0]}")  #Mostra o número de linhas do DataFrame
print(f"Colunas: {df.shape[1]}") #Mostra o número de colunas do DataFrame
print(f"\nNomes das colunas:")   #Mosra o título "Nomes das colunas:"
print(df.columns.tolist())       #Mostra os nomes das colunas do DataFrame em formato de lista
```

```python
# Célula 3 — Linha específica pelo índice
print("Linha específica pelo índice (exemplo: índice 3):\n")
df.iloc[3]
```

```python
#  Célula 4 — tipos e nulos

# .info() — resumo completo: tipos, nulos e uso de memória
# É o comando mais importante para um primeiro diagnóstico
print("Resumo completo do DataFrame:\n")
df.info()
```

```python
# Quantidade de nulos por coluna
print("\nQuantidade de valores nulos por coluna:\n")
print("\nNulos por coluna:")
print(df.isnull().sum())

# Percentual de nulos — mais útil em bases grandes
print("\n% de nulos:")
print((df.isnull().sum() / len(df) * 100).round(2))
```

```python
# Célula 5 — estatísticas numéricas

# .describe() — estatísticas das colunas numéricas
# count, mean, std, min, 25%, 50%, 75%, max
df.describe().round(2)
```

```python
# Célula 6 — cardinalidade e distribuição

# Valores únicos por coluna — identifica colunas categóricas
print("Valores únicos por coluna:\n")
print(df.nunique())

# Distribuição de cada coluna categórica relevante
for col in ("Departamento", "Cargo", "Genero", "Status", "Estado_Civil"):
    print(f"\n── {col} ──")
    print(df[col].value_counts())
```

```python
# Célula 7 — seleção de colunas e filtros simples

# Selecionar uma coluna → Series
df["Nome"].head(3)
```

```python
# Selecionar múltiplas colunas → DataFrame
df[["Nome", "Departamento", "Salario"]].head(4)
```

```python
# Filtro simples: apenas funcionários Ativos
ativos = df[df["Status"] == "Ativo"]
inativos = df[df["Status"] == "Inativo"]
print(f"Ativos: {len(ativos)} | Inativos: {len(inativos)}")
```

```python
# Filtro composto: Gerentes do departamento de TI
gerentes_ti = df[(df["Cargo"] == "Gerente") & (df["Departamento"] == "TI")]
print(f"\nGerentes de TI:")
print(gerentes_ti[["Nome", "Salario", "Status"]].to_string(index=False))
```

```python
# Estatísticas rápidas de uma coluna
print(f"\nSalário — média: R$ {df['Salario'].mean():,.2f}")
print(f"Salário — maior: R$ {df['Salario'].max():,.2f}")
print(f"Salário — menor: R$ {df['Salario'].min():,.2f}")
```

**Por que isso importa ?**

Esses 7 comandos — shape, head, tail, info, isnull, describe, nunique + value_counts — são o protocolo mínimo de qualquer analista antes de tocar nos dados. 
Em mercado, você nunca entrega uma análise sem ter rodado esses comandos primeiro: eles revelam nulos escondidos, tipos errados e distribuições inesperadas que invalidam qualquer conclusão posterior.
